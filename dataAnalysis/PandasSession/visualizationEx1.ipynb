```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nums = np.arange(1,11)
sq = nums ** 2
```

```python
plt.plot(nums,sq, marker='o')
plt.xlabel("Numbers")
plt.ylabel("squares")
plt.title("line chart")
plt.show()
```

```python
plt.plot(nums, sq, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.savefig("linechar.jpg")
```
