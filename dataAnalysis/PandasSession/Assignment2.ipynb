```python
import pandas as pd
import numpy as np
```

### 1. Create a DataFrame:
#### o Load the above dataset into a Pandas DataFrame.

```python
dataset = pd.read_csv("datasets/sales.csv")
dataset
```

### 2. Add a New Column:
#### o Add a column Discount that gives a 10% discount on total sales.

```python
dataset['Discount'] = dataset['Total Sales'] - dataset['Total Sales'] * 0.1
```

```python
dataset
```

### 3. Add a New Row:
#### o Add a new sales record: Order ID = 1009, Customer ="Eve", Product = "Tablet", Category = "Electronics",
#### Quantity = 1, Price = 300, Total Sales = 300.

```python
dataset.loc[len(dataset)] = [1009,"Eve",'Tablet','Electronics',1,300,300,270]
```

```python
dataset
```

### 4. Filter Data:
#### o Retrieve only the sales records where the Category is "Electronics".

```python
dataset[dataset['Category'] == 'Electronics']
```

### 5. Selection:
#### o Select the Customer and Total Sales columns.

```python
dataset[['Customer','Total Sales']]
```

### 6. Grouping:
#### o Group the data by Customer and calculate the total sales per customer.

```python
dataset.groupby('Customer')['Total Sales'].sum()
```

### 7. Sorting:
#### o Sort the DataFrame based on Total Sales in descending order.

```python
dataset.sort_values(by='Total Sales',ascending=False)
```

### 8. Apply Function:
#### o Apply a lambda function to create a new column Sales
#### Category, where values are labeled as "High" if Total Sales > 1000, otherwise "Low".

```python
dataset['Sales Category'] = dataset['Total Sales'].apply(lambda a: "High" if a >= 1000 else "Low")
```

```python
dataset
```

### 9. Save DataFrame:
#### o Save the final DataFrame as a CSV file named sales_data.csv.

```python
dataset.to_csv("sales_data.csv", index=False) 
```
